```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

importation du dataframe

```python
data = pd.read_excel('./' + 'Coeur.xlsx')
df_1 = data.copy()
```

suppression des doublons

```python
df_1 = df_1.drop_duplicates().reset_index(drop=True)
df_1
```

verification des donnees manquantes

```python
df_1.dropna()
```

supression des constantes

```python
df_1.loc[:, (df_1 != df_1.iloc[0]).any()] 
#data = data.loc[:,data.apply(pd.Series.nunique) != 1]
#data
```

recodage des variables qualitatives

```python
def record_serie(serie):
    return serie.astype('category').cat.codes

def record_dataframe(df):
    for col in df.select_dtypes('object').columns:
        df[col]= record_serie(df[col])
    return df
df_1=record_dataframe(df_1)
df_1
```

normalisation

```python
var_quanti = df_1.select_dtypes('int')
var_quanti
```

```python
def normalize_quat(df):
    for i in df.select_dtypes('int'):
        df[i]=df[i]/df[i].max()
        return df

normalize_quat(df_1)
```

```python
df_1 = df_1.fillna(df_1.mean())
```

attribuons a x [age,pente] et a y[coeur]

```python
x = df_1.iloc[:,:-1]
y = df_1.iloc[:,11:12]
x
```

```python
model = KNeighborsClassifier() #importation de notre model de classification
model.fit(x, y) # pour faire l'entrainement du model
model.score(x, y) #pour voir le score de notre model
```

```python
model.predict(x)#permet de predire notre model
```
